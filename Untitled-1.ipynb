```python
import pandas as pd
import re

# Định nghĩa mẫu Regex để bóc tách đúng 10 cột như bạn yêu cầu
# Sử dụng Named Groups (?P<tên_cột>) để Pandas tự nhận diện tên cột
log_pattern = r'(?P<IP>\S+) (?P<Remote_user>\S+ \S+) \[(?P<Time>.*?)\] "(?P<Method>\S+) (?P<Path>.*?) (?P<Protocol>HTTP/\d\.\d)" (?P<Status>\d{3}) (?P<Size>\S+) "(?P<Referrer>.*?)" "(?P<User_Agent>.*?)"'

def parse_logs_to_dataframe(file_path):
    data = []
    
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if not line: continue
            
            match = re.search(log_pattern, line)
            if match:
                data.append(match.groupdict())
                
    # Tạo DataFrame từ danh sách các dictionary đã bóc tách
    df = pd.DataFrame(data)
    
    # Ép kiểu dữ liệu cho các cột số để tiện xử lý sau này
    if not df.empty:
        df['Status'] = pd.to_numeric(df['Status'], errors='coerce')
        df['Size'] = pd.to_numeric(df['Size'].replace('-', '0'), errors='coerce')
        
    return df

# --- CHẠY THỬ ---
# df = parse_logs_to_dataframe('your_log_file.log')

# Demo với dòng dữ liệu của bạn
sample = '54.36.149.41 - - [22/Jan/2019:03:56:14 +0330] "GET /filter/27|13%20%D9%85%DA%AF%D8%A7%D9%BE%DB%8C%DA%A9%D8%B3%D9%84 HTTP/1.1" 200 30577 "-" "Mozilla/5.0 (compatible; AhrefsBot/6.1; +http://ahrefs.com/robot/)"'

match = re.search(log_pattern, sample)
if match:
    df_result = pd.DataFrame([match.groupdict()])
    # Hiển thị danh sách các cột để bạn kiểm tra
    print("Danh sách các cột trong DataFrame:")
    print(df_result.columns.tolist())
    print("\nDữ liệu chi tiết:")
    print(df_result.iloc[0])
```

```python
import polars as pl

# Giả sử cột user agent của mày tên là 'user_agent'
# Tạo một cột mới 'browser' dựa trên từ khóa trong user_agent
browser_df = df.with_columns(
    pl.when(pl.col("user_agent").str.contains("(?i)chrome"))
    .then(pl.lit("Chrome"))
    .when(pl.col("user_agent").str.contains("(?i)firefox"))
    .then(pl.lit("Firefox"))
    # Lưu ý: Chrome cũng chứa chữ 'Safari', nên phải check Chrome trước hoặc check kỹ hơn
    .when(pl.col("user_agent").str.contains("(?i)safari") & ~pl.col("user_agent").str.contains("(?i)chrome"))
    .then(pl.lit("Safari"))
    .otherwise(pl.lit("Others"))
    .alias("browser")
)

# Đếm số lượng
browser_counts = browser_df.group_by("browser").agg(pl.count()).sort("count", descending=True)

print(browser_counts)
print("-------------------------------------------------------------------------------------------------------")
import matplotlib.pyplot as plt

# 1. Chuyển kết quả từ Polars sang Pandas để vẽ hình cho mượt
data = browser_counts.to_pandas()

# 2. Cấu hình biểu đồ
plt.figure(figsize=(10, 6))

# Định nghĩa màu sắc cho từng trình duyệt (tùy chọn)
colors = {
    'Chrome': '#4285F4',   # Xanh Google
    'Firefox': '#FF7139',  # Cam Cáo
    'Safari': '#00AEEF',   # Xanh Blue Safari
    'Others': '#9E9E9E'    # Xám cho bọn còn lại
}

# Ánh xạ màu theo tên trình duyệt trong dữ liệu
bar_colors = [colors.get(name, '#34A853') for name in data['browser']]

# 3. Vẽ biểu đồ cột
bars = plt.bar(data['browser'], data['count'], color=bar_colors, edgecolor='black', alpha=0.8)

# 4. Thêm số liệu cụ thể trên đầu mỗi cột cho dễ đọc
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + (yval*0.01), 
             f'{int(yval):,}', ha='center', va='bottom', fontweight='bold')

# 5. Trang trí
plt.title("The number of requests in different browsers", fontsize=15, pad=20)
plt.xlabel("Browsers", fontsize=12)
plt.ylabel("Requests", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7) # Thêm lưới ngang cho dễ nhìn

plt.tight_layout()
plt.show()
```

#### Finding abnormal samples which are reflected through their sizes and request times

In this project, any samples had been abnormal if they were rarely requested but they accounted for over 1 GB

```python
monsters = (
    df.group_by("ip")
    .agg([
        pl.count().alias("request_count"),
        ((pl.col("size").sum()/(1024**3)).alias("total_gb"))
    ])
    .filter(pl.col("total_gb") > 1)
)
print("-------------------------------------------------------------------------------------------------------")
print("List of IPs that downloaded more than 1 GB of data: ")
monsters
```

## Identify the "attacking" IPs (100 requests/min)

```python
import polars as pl

# Vì cột 'time' đã là kiểu datetime[μs, UTC], ta dùng trực tiếp luôn
df = df.with_columns(
    pl.col("time").dt.truncate("1m").alias("minute")
)



# 1. Gom nhóm theo cả phút và IP
ip_spam_check = (
    df.group_by(["minute", "ip"])
    .agg(pl.count().alias("request_count"))
)

# 2. Lọc ra những ông nào có request_count > 100
spam_list = (
    ip_spam_check
    .filter(pl.col("request_count") > 100)
    .sort("request_count", descending=True)
)

print("List of IPs that made more than 100 requests in a single minute: ")
print(spam_list)
```
